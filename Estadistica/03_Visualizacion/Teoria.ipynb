## Diagrama de Dispersión

Un gráfico de dispersión o diagrama de dispersión es un tipo de diagrama en el que se indican los valores individuales
usando coordenadas cartesianas (o x e y) para mostrar valores. Cada valor se indica
mediante una marca (como un círculo o un triángulo) en el gráfico. Se pueden usar para representar valores obtenidos para dos variables diferentes; uno trazado en el eje x y el otro trazado en el eje y.

````python
# Crearemos los datos
## Haremos numero de ejercicios que realiza un alumno y horas practicando la materia del curso
python = ((17, 18, 21, 22, 19, 21, 25, 22, 25, 24),
(3, 6, 3.5, 4, 5, 6.3, 4.5, 5, 4.5, 4))
sql = ((17, 18, 20, 19, 22, 21, 23, 19, 21, 24),
(8, 9, 7, 10, 7.5, 9, 8, 7, 8.5, 9))
matematicas = ((31, 28, 29, 36, 27, 32, 34, 35, 33, 39),
(4, 6.3, 6, 3, 5, 7.5, 2, 5, 7, 4))
# Representamos los datos
plt.scatter(x=python[0], y=python[1], c='red', marker='o',
label='python')
plt.scatter(x=sql[0], y=sql[1], c='green',
marker='^', label='sql')
plt.scatter(x=matematicas[0], y=matematicas[1], c='blue',
marker='*', label='matematicas')
#configuración del gráfico
plt.xlabel('Ejercicios')
plt.ylabel('Horas')
plt.title('Diagrama de Dispersión de Ejercicios')
plt.legend()
plt.show();
````

### ¿Siempre se debe usar un diagrama de dispersión?
El diagrama de dispersión permite estudiar las relaciones entre dos conjuntos asociados de datos que aparecen en pares (por ejemplo, (x,y), uno de cada conjunto). El diagrama muestra estos pares como una nube de puntos.

Las relaciones entre los conjuntos asociados de datos se infieren a partir de la forma de las nubes.

* Una relación positiva entre x y y significa que los valores crecientes de x están asociados con los valores crecientes de y.
* Una relación negativa significa que los valores crecientes de x están asociados con los valores decrecientes de y. 

Para facilitar esa relación o tendencia se utilizará de la librería de ```numpy``` la función ```polyfit()```.

````python
import numpy as np
x = (5, 5.5, 6, 6.5, 7, 8, 9, 10)
y = (120, 115, 100, 112, 80, 85, 69, 65)
# Generamos el diagrama de dispersión
plt.scatter(x, y)
# Generamos la línea de tendencia
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), 'r')
# Mostramos el gráfico
plt.show()
````

## Diagrama de Sectores

Un diagrama de sectores (más comúnmente conocido como diagrama de tarta) es un círculo se divide en sectores donde cada uno representa una proporción del todo. Un sector del círculo representa una contribución de la categoría al total general. Como tal, el gráfico se asemeja a un pastel que se ha cortado en rodajas de diferentes tamaños.
Por lo general, los diferentes sectores del gráfico circular se presentan en diferentes colores y están dispuestos en el sentido de las agujas del reloj alrededor del gráfico en orden de magnitud. Sin embargo, si hay una porción que no contiene una categoría única de datos pero resume
varios, por ejemplo, "otros tipos" u "otras respuestas", incluso si no es el categoría más pequeña, es habitual mostrarlo en último lugar para que no desmerezca las categorías de interés nombradas.

````python
import matplotlib.pyplot as plt
labels = ('Python', 'PBI', 'SQL', 'Tableau')
Porcentajes = [45, 30, 15, 10]
plt.pie(Porcentajes,
        labels=labels,
        autopct='%1.f%%',
        counterclock=False,
        startangle=90)
plt.show();
````

La función `pyplot.pie()` toma muchos parámetros, muchos de los cuales son opcionales. El único parámetro requerido es el primero que proporciona los valores a ser utilizado para los tamaños de los segmentos. 
Los siguientes parámetros opcionales en el ejemplo han sido:

* El parámetro ``label`` es un opcional y puede tomar una secuencia de strings que se utulizan para las eitquetas de cada de los sectores.
* El parámetro ``autopct`` toma string (o funciones) para usar un formato numérico de cada uno de los sectores.
* El parámetro ``counterclockwise``. Por defecto los sectores son dibujados en el sentido contratio de las agujas del reloj. Para asegurarnos de que el diseño se parece al tradicional, el parámetro de ``counterclockwise`` se establece en `Falso`
* El parámetro ``startangle``. El ángulo inicial también se ha movido 90° utilizando el ángulo de inicio

#### Desplegando Segmentos
Para mostrar más ***enfasis*** en un segmento en particular del gráfico de segmentos, este segmento se separa del resto de segmentos. Esto se puede hacer usando el parámetro ``explode`` que tomará una secuencia de valores que indica cuanto debe separarse.

El impacto visual del gráfico circular también se puede mejorar en este caso agregando una _sombra_ a los segmentos usando el parámetro booleano ``shadow``.

````python
labels = ('Python', 'PBI', 'SQL', 'Tableau')
Porcentajes = [45, 30, 15, 10]
Separacion = (0.1, 0, 0, 0) ## se va a resaltar el sector de python
plt.pie(Porcentajes,
        labels=labels,
        explode=Separacion,
        autopct='%1.f%%',
        shadow=True,
        counterclock=False,
        startangle=90)
plt.show();
````

## Gráfico de Barras

Un gráfico de barras es un tipo de gráfico que se utiliza para presentar diferentes categorías de datos. Los datos se suelen presentar de forma vertical aunque en algunos casos se pueden utilizar gráficos de barras horizontales. Cada categoría está representada por una barra cuyo
altura (o longitud) representa los datos para esa categoría.
Debido a que es fácil interpretar los gráficos de barras y cómo cada categoría se relaciona con otra, son uno de los tipos de gráficos más utilizados. También hay varios diferentes variaciones comunes, como gráficos de barras agrupadas y gráficos de barras apiladas.
El siguiente es un ejemplo de un gráfico de barras típico. Utilizando los datos del [ejemplo anterior](#diagrama-de-sectores) los idiomas se presentan a lo largo del eje x, mientras que el eje y indica porcentaje de uso. Cada barra representa el porcentaje de uso asociado con
cada lenguaje de programación.

````python

labels = ('Python', 'PBI', 'SQL', 'Tableau')
indice = (1, 2, 3, 4) 
Porcentajes = [45, 10, 15, 30]

plt.bar(indice, Porcentajes, tick_label=labels)

plt.ylabel('Porcentaje')
plt.xlabel('Lenguaje')

plt.show();
````

El gráfico está construido de manera que las longitudes de las diferentes barras sean proporcionales al tamaño de la categoría que representan. El eje x representa los diferentes categorías y por lo tanto no tiene escala. Para enfatizar el hecho de que las categorías son
discreta, se deja un espacio entre las barras en el eje x. El eje y tiene una escala y esto indica la unidad de medida.

Muy **recomendable** que para una comparación más rapida se ordene las categorías haciendo que los porcentajes estén ordenados de mayor a menor o viceversa. 

### Gráfico de Barras Horizontal

Los gráficos de barras normalmente se dibujan de modo que las barras sean verticales, lo que significa que la barra más alta, mayor frecuencia tiene la categoría. Sin embargo, también es posible dibujar gráficos de barras de modo que las barras queden horizontales, lo que significa que cuanto más larga sea la barra, mayor frecuencia tiene la categoría. Esta es una manera particularmente eficaz de presentar un gran número de
diferentes categorías cuando no hay suficiente espacio para acomodar todas las columnas requeridas para un gráfico de barras verticales a lo largo de la página.
En Matplotlib, la función ``pyplot.barh()`` se puede usar para generar una horizontal gráfico de barras:

````python
labels = ('Python', 'PBI', 'SQL', 'Tableau')
indice = (1, 2, 3, 4) 
Porcentajes = [45, 10, 15, 30]
plt.barh(indice, Porcentajes, tick_label = labels)
plt.show();
````

### Barras de colores

Es común colorear de diferente color cada una de las barras para así distinguir una barra de otra.

El color que se usa para cada una de las cateogrías viene vía por el parámetro ``color`` de la función ``bar()`` o ``barh()``
Se crea una secuencia de colores

````python
labels = ('Python', 'PBI', 'SQL', 'Tableau')
indice = (1, 2, 3, 4) 
Porcentajes = [45, 10, 15, 30]
colores=("blue","red","green","orange")
plt.bar(indice, Porcentajes, tick_label = labels,color=colores)
plt.show();
````

### Gráfico de Barras Apiladas
Los gráficos de barras también se pueden apilar. Esta puede ser una forma de mostrar los valores totales (y lo que
contribuye a esos valores totales) en varias categorías. Es decir, es una forma de
ver los totales generales, para varias categorías diferentes en función de cómo los diferentes elementos
contribuyen a esos totales.
Se utilizan diferentes colores para los diferentes subgrupos que contribuyen a la
barra general. En tales casos, se suele proporcionar una leyenda o clave para indicar qué
subgrupo que representan cada uno de los matices/colores.

````python
labels = ('Python', 'PBI', 'SQL', 'Tableau','Matematicas')
indice = (1, 2, 3, 4, 5)
programador = [20, 2, 5, 10, 5]
data_science = [15, 8, 5, 15, 15]
data_analisis = [10, 1, 5, 5, 10]

plt.bar(indice, programador, tick_label=labels, label='Programador')
plt.bar(indice, data_science, tick_label=labels,
label='Data science', bottom=programador)
web_science = [programador[i] + data_science[i]
for i in range(0, len(programador))]
plt.bar(indice, data_analisis, tick_label=labels,
label='Data analisis', bottom=web_science)

plt.ylabel('Rama')
plt.xlabel('Lenguaje')
plt.legend()

plt.show()
````

### Gráfico de Barras Agrupadas
Por último, los gráficos de barras agrupadas son una forma de mostrar información sobre diferentes subgrupos de las categorías principales. En tales casos, se suele proporcionar una leyenda o clave para indicar qué subgrupo representa cada uno de los matices/colores. La leyenda se puede colocar en el área de trazado o se puede ubicar debajo del gráfico.

Para una categoría en particular, se dibujan gráficos de barras separados para cada uno de los subgrupos.

Se muestra la puntuación de dos usuarios en cinco pruebas que se han realizado 

````python
Achura_Barra = 0.35

Usuario_A = (60, 75, 56, 62, 58)
Usuario_B = (55, 68, 80, 73, 55)

Indice_A = (1, 2, 3, 4, 5)
Indice_B = [i + Achura_Barra for i in Indice_A]

ticks = [i + Achura_Barra / 2 for i in Indice_A]
tick_labels = ('Prueba 1', 'Prueba 2', 'Prueba 3', 'Prueba 4', 'Prueba 5')

plt.bar(Indice_A, Usuario_A, Achura_Barra, color='b',
label='Usuario A')
plt.bar(Indice_B, Usuario_B, Achura_Barra, color='g',
label='Usuario B')

plt.xlabel('Prueba')
plt.ylabel('Puntuacion')
plt.title('Puntuación por Prueba')
## con el parametro rotation podemos rotar los xticks un cierto ángulo
plt.xticks(ticks, tick_labels,rotation=45)
plt.legend()
plt.show()
````

## Figuras y Subgráficos
Una figura de Matplotlib es el objeto que contiene todos los elementos gráficos que se muestran
como puede ser:  los ejes, la leyenda, el título y el gráfico.


La función ``matplotlib.pyplot.figure()`` genera un objeto de figura.
Esta función devuelve un objeto ``matplotlib.figure.Figure.`` entonces es posible para interactuar directamente con el objeto figura. Por ejemplo, es posible agregar ejes a la figura, para agregar subgráficos a un gráfico, etc.

Es necesario trabajar directamente con la figura si desea agregar varias subgráficos a una figura Esto puede ser útil si lo que se requiere es poder comparar diferentes vistas de los mismos datos una al lado de la otra. Cada subgráfico tiene sus propios ejes que pueden coexistir dentro de la figura.

Se pueden agregar uno o más subgráficos a una figura usando ``figure.add_subplot()``. 

Este método agrega ejes a la figura como uno de un conjunto de uno o más subgráficos. Se puede agregar un subgráficos usando un número entero de 3 dígitos (o tres enteros) que describen la posición del subgráficos. Los dígitos representan el número de filas, columnas y el índice del subgráficos dentro de la matriz resultante.

Por lo tanto, 2, 2, 1 (221) indican que el subgráfico tomará el primer índice dentro de un cuadrícula de de dos por dos. A su vez 2, 2, 3 (223) indica el subgráfico estará en índice 3, que será la fila 2 y la columna 2 dentro de la cuadrícula de 2 por 2. Mientras 2, 2, 4 (224) indica el subgráfico debe agregarse como en el índice 4 o el cuarto el subgráfico dentro de la cuadrícula posición 2 por 2, etc.

````python
t = range(0, 20)
s = range(30, 10, -1)

grid_size='22'

figure = plt.figure()

position = grid_size + '1'
print('Añadiendo el primero subplot a la posicion', position)
axis1 = figure.add_subplot(int(position))
axis1.set(title='subplot(2,2,1)')
axis1.plot(t, s)

position = grid_size + '2'
print('Añadiendo el primero subplot a la posicion', position)
axis2 = figure.add_subplot(int(position))
axis2.set(title='subplot(2,2,2)')
axis2.plot(t, s, 'r-')

position = grid_size + '3'
print('Añadiendo el primero subplot a la posicion', position)
axis3 = figure.add_subplot(int(position))
axis3.set(title='subplot(2,2,3)')
axis3.plot(t, s, 'g-')

position = grid_size + '4'
print('Añadiendo el primero subplot a la posicion', position)
axis4 = figure.add_subplot(int(position))
axis4.set(title='subplot(2,2,4)')
axis4.plot(t, s, 'y-')

plt.show()
````

## Gráficos en 3D
Se utiliza un gráfico tridimensional para trazar las relaciones entre tres conjuntos de en lugar de los dos utilizados en los ejemplos. 
En un gráfico tridimensional, además de los ejes x e y, también hay un eje z.

Para crea un gráfico 3D simple usando dos conjuntos de valores generado usaremos la función de rango ``numpy``. 
Estos luego se convierten en coordenadas de matrices usando la función ``numpy`` ``meshgrid()``. 
Los valores del eje z son creados usando la función ``numpy`` ``sin()``. 
La superficie del gráfico 3D se traza usando el función ``plot_surface()`` del objeto de ejes de futuros. 
Esto toma la x, y y z coordenadas La función también recibe un mapa de color para usar al renderizar el superficie.


````python
from matplotlib import cm as colourmap
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
x_values = np.arange(-6, 6, 0.3)
y_values = np.arange(-6, 6, 0.3)
X, Y = np.meshgrid(x_values, y_values)
Z = np.sin(X + Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap=colourmap.coolwarm)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

plt.title("3D Grafico")
plt.show();
````

````python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap=colourmap.coolwarm)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
````
