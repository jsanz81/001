```python
amapola.drop(['count', 'min','max'], axis=1, inplace=True)
# columnas que no queremos
```

```python
listacurtosis=[]
listaasimetria=[]
listarango=[]
listacoefvariacion=[]

# listas con medidas pedidas
```

```python
indice=amapola.index
# lista de productos
```

```python
for i in indice:
    listacurtosis.append(ss.kurtosis(super[i]))
    listaasimetria.append(ss.skew(super[i]))
    listarango.append(ss.iqr(super[i]))
    listacoefvariacion.append(super[i].std()/super[i].mean())
```

```python
amapola["curtosis"]=listacurtosis
amapola["asimetria"]=listaasimetria
amapola["rangoIQR"]=listarango
amapola["coefVariacion"]=listacoefvariacion
#añadir listas al dataframe
```


* si la distribución del producto es asimétrica por la derecha y platicúrtica el precio será: 
* la tercera parte del rango intercuartílico por el coeficiente de variación

* si la distribución del producto es asimétrica por la izquierda y platicúrtica
* el precio será: la media por la décima parte de la desviación típica

* si la distribución del producto Y es asimétrica por la derecha y leptocúrtica
* el precio será: la mediana por el coeficiente de asimetría en valor absoluto.

* en otros casos será diez veces el percentil 50


```python
def funcionPrecio(s):
    if (s['asimetria']>0 and s['curtosis']<0):
        precio=(s['rangoIQR']/3)*s['coefVariacion']
        return(precio)
    elif (s['asimetria']<0 and s['curtosis']<0):
        precio=s['mean']*(s['std']/10)
        return(precio)
    elif (s['asimetria']>0 and s['curtosis']>0):
        precio=s['50%']*abs(s['asimetria'])
        return(precio)
    else:
        precio=s['50%']*10
        return(precio)
```

```python
amapola["precio"]=amapola.apply(lambda x: funcionPrecio(x), axis=1)
```
