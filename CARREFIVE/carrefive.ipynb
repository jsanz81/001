```python
import pandas as pd
import numpy as np
import sqlite3
import os
```

```python
# Creamos la conexi√≥n para DB local

conn = sqlite3.connect("carrefive.db")

# Creamos el comando o cursor
cursor = conn.cursor()
```

```python
tablas_de_bd = pd.read_sql(""" select *
                        from sqlite_master 
                        where type='table'
                        order by name;
                        """, conn)
tablas_de_bd
```

**1.** Selecciona los clientes de la India y Brasil

```python
q = """
    SELECT customer_id, country FROM CLIENTE
    WHERE country IN ("IN", "BR")
"""
df = pd.read_sql(q, conn)
df
```

**2.** Selecciona los clientes italianos que no se encuentren en las categor√≠as 4 y 5

```python
q = """
    SELECT customer_id, tier
    FROM CLIENTE
    WHERE (tier LIKE '1%') OR (tier LIKE '2%') OR (tier LIKE '3%')
"""
df = pd.read_sql(q, conn)
df
```

**3.** Selecciona los clientes europeos que hablen castellano

```python
q = """
    SELECT DISTINCT country
    FROM CLIENTE
"""
df = pd.read_sql(q, conn)
df.values
```

```python
q = """
    SELECT customer_id, country, spanish_speaker
    FROM CLIENTE
    WHERE
        country IN ('AU','IT','FR','SP','DE','UK') AND spanish_speaker=TRUE
        
"""
df = pd.read_sql(q, conn)
df
```

**4.** Averigua la nacionalidad que m√°s clientes tenga a los que se les puede enviar e-mails

```python
q = """
    SELECT c.country, COUNT(c.customer_id) as "N¬∫ Clientes del pais", permiso_email
    FROM CLIENTE as c, MARKETING as m
    WHERE
        c.customer_id=m.customer_id AND
        PERMISO_EMAIL="Y"
    GROUP BY c.country
    ORDER BY 2 DESC
"""
df = pd.read_sql(q, conn)
df
```

**5.** Numero total de transacciones que consiguen puntos

```python
q = """
    SELECT COUNT(tipo_transaccion) as "Transacciones con Puntos"
    FROM OBTENCION_PUNTOS
"""
df = pd.read_sql(q, conn)
df
```

**6.** N√∫mero total de puntos conseguidos 

```python
q = """
    SELECT SUM(puntos_conseguidos) as "Puntos Conseguidos"
    FROM OBTENCION_PUNTOS
"""
df = pd.read_sql(q, conn)
df
```

**7.** N√∫mero total de clientes distintos que han utilizado puntos

```python
q = """
    SELECT DISTINCT(COUNT(customer_id)) as "Clientes que han usado puntos"
    FROM USO_DE_PUNTOS
"""
df = pd.read_sql(q, conn)

df
```

**8.** N√∫mero total de puntos utilizados

```python
q = """
    SELECT SUM(puntos_utilizados) as "Suma de Puntos"
    FROM USO_DE_PUNTOS
"""
df = pd.read_sql(q, conn)
df
```

**9.** Proporci√≥n de transacciones en las que se obtienen puntos al comprar pescado con respecto al total de transacciones en las que se obtienen puntos

```python
q = """
    SELECT *
    FROM USO_DE_PUNTOS
"""
df = pd.read_sql(q, conn)
df
```

```python
q = """
    SELECT COUNT(customer_id)
    FROM OBTENCION_PUNTOS
    WHERE obtencion_puntos="Pescado"
"""
df = pd.read_sql(q, conn)
df
```

**10.** Proporci√≥n de transacciones en las que se obtienen puntos al comprar pescado con respecto al total de transacciones

```python
q = """
    SELECT obtencion_puntos
    FROM OBTENCION_PUNTOS
    WHERE obtencion_puntos="Pescado"
"""

puntos_pescado = pd.read_sql(q, conn).shape[0]

print(puntos_pescado)

q2 = """
    SELECT obtencion_puntos
    FROM OBTENCION_PUNTOS
"""
puntos_total = pd.read_sql(q2, conn).shape[0]
print(puntos_total)

print(f"Proporci√≥n puntos Pescado: {puntos_pescado/puntos_total*100}%")
```
