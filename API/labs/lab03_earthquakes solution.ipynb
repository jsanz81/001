```python
import datetime as dt
import pandas as pd
import requests
```

```python
# los últimos 15 días
two_weeks = dt.date.today() - dt.timedelta(days=15)

ayer = dt.date.today() - dt.timedelta(days=1)
```

* https://earthquake.usgs.gov/fdsnws/event/1/

* web de la API

* https://earthquake.usgs.gov/fdsnws/event/1/[METHOD[?PARAMETERS]]

```python
url = 'https://earthquake.usgs.gov/fdsnws/event/1/query'
starttime = two_weeks
endtime = ayer
```

```python
payloads = {
    'format' : 'geojson',
    'starttime' : starttime,
    'endtime' : endtime
}
```

```python
# Creo la solicitud
response = requests.get(url=url, params=payloads)
response.status_code
```

```python
response.json()
```

```python
terremotos = response.json()
terremotos.keys()
```

```python
terremotos['features']
```

```python
terremotos_dict = [
                   quake['properties'] for quake in terremotos['features']
]
df = pd.DataFrame(terremotos_dict)
df.head()
```
